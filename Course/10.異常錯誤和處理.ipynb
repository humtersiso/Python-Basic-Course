```python
try :
    # 執行你要測試的代碼
    a = [12, 3, 4, 67, 55, 0, 12, 4]
    b = [i for i in a if 2 % i == 0]
    c = [i for i in a if i % 3 == 0]

except :
    # 如果發生錯誤了，執行以下代碼
    print('有錯誤發生了')    
```

```python
try :
    # 執行你要測試的代碼
    a = [12, 3, 4, 67, 55, 0, 12, 4]
    b = [i for i in a if 2 % i == 0]
    c = [i for i in a if i % 3 == 0]

except ZeroDivisionError :
    # 如果發生錯誤了，執行以下代碼
    print('0不能是被除數')   
```

#### d. 錯誤處理的其他用法

```python
try:
    # 執行你要測試的代碼

except:
    # 如果發生錯誤了，執行以下代碼

else:
    # 沒有發生錯誤時執行下列代碼
    
```

```python
list1 = [1, 2, 3, '沙阡陌']
list2 = [4, 5, 6]
```

```python
try:
    print(sum(list1))

except:
    list1[3] = 0
    print(sum(list1 + list2))

else:
    print(sum(list1 + list2))
```

```python
try:
    # 執行你要測試的代碼

except:
    # 如果發生錯誤了，執行以下代碼

else:
    # 沒有發生錯誤時執行下列代碼

finally:
    # 不管有沒有發生錯誤都會執行
```
