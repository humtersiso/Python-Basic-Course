#### b. 異常錯誤處理

```python
try:
    ##執行你要測試的代碼
    a = [12 ,3 ,4 ,67 ,55 ,0 , 12 ,4 ]
    b = [i for i in a if 2 % i ==0 ]

except:
    ##如果發生錯誤了，執行以下代碼
    print('有錯誤發生了')
    

print('程序結束')
```

```python
try:
    a = [12 ,3 ,4 ,67 ,55 ,0 , 12 ,4 ]
    b = [i for i in a if 2 % i ==0 ]

except ZeroDivisionError:
    print('0不能是被除數')
    

print('程序結束')
```

```python
try:
    a = [12 ,3 ,4 ,67 ,55 ,1 , 12 ,4 ,'胡萬']
    b = [i for i in a if 2 % i ==0 ]

except ZeroDivisionError:
    print('0不能是被除數')
    

print('程序結束')
```

```python
try:
    a = [12 ,3 ,4 ,67 ,55 ,1 , 12 ,4 ,'胡萬']
    b = [i for i in a if 2 % i ==0 ]

except ZeroDivisionError:
    print('0不能是被除數')

except TypeError:
    print('列表裡有非整數的元素')

print('程序結束')
```

#### c. 用戶自定義錯誤

```python
raise NameError('名稱錯誤')
```

```python
raise TypeError('數據類型錯誤')
```

```python
try:
    raise NameError('名稱錯誤')
except NameError:
    print('name error')
```

```python
try:
    raise NameError('名稱錯誤')
except NameError as ne:
    print(ne)
```

```python
try:
    raise NameError('名稱錯誤')
except TypeError as ne:
    print(ne)
```

#### d. 錯誤處理的其他用法

```python
try:
    ##執行你要測試的代碼

except:
    ##如果發生錯誤了，執行以下代碼

else:
    ##沒有發生錯誤時執行下列代碼
    
```

```python
list1 = [1 ,2 ,3 ,'沙阡陌']
```

```python
list1 = [1 ,2 ,3 ,'沙阡陌']
list2 = [4 ,5 ,6]

try:
    print(sum(list1)) #測試這段代碼
    
except:
    list1[3] =0
    print(sum(list1 + list2 ))

else:
    print(sum(list1 + list2 ))            
```

```python
list1 = [1 ,2 ,3 ,'沙阡陌']
list2 = [4 ,5 ,6]

try:
    print(sum(list1)) #測試這段代碼
    
except:
    list1[3] =0
    print(sum(list1 + list2 ))

else:
    print(sum(list1 + list2 )) 

finally:
    print('計算完成')
```
