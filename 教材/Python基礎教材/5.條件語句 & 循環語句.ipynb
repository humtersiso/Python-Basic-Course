#### a.條件語句
  * if...語句

```python
temp = 32
if temp >30:
    print('好熱，快開冷氣')
```

* if...else...語句

```python
temp = 32
if temp > 35 :
    print('好熱，快開冷氣')
else :
    print('涼爽~涼爽~')
```

* if...elif...else...語句

```python
temp = 22
if temp > 30 :
    print('好熱，快開冷氣')
elif temp >28 and temp<=30:
    print('有點熱耶!')
elif temp >20 and temp<=28:
    print('涼爽~涼爽~')
else :
    print('好像有點冷!')
```

```python
temp = -22
if (temp>= -80 and temp <= -20) or (temp >=35 and temp <= 50):
    print('極端的溫度')
else :
    print(None)
```

#### b.循環語句(while)

* 基本概念

```python
temp = 28
while temp < 40:
    print('溫度:',temp ) #什麼時候會停止死循環?
```

```python
temp = 28
while temp < 40:
    print('溫度:',temp )
    temp = temp + 1
    #temp +=1
```

```python
temp
```

```python
temp = 28
while temp < 40:
    print('溫度:',temp )
    temp +=1     #temp需要重新指定
```

* 和條件語句的搭配

```python
temp = 28
while temp < 40:
    print('溫度:',temp)
    temp +=1
    if temp >32: 
        print('有點熱耶!')
```

```python
temp = 28
while  temp < 30:
    print('溫度:',temp,'涼爽~涼爽~')
    temp +=1
else :
    print('溫度:',temp,'好熱，快開冷氣')
```

```python
turn_off=[]
turn_on=[]
temp = 28

while temp < 38:
    temp +=1   
    if temp >30: 
        turn_on.append(temp)
    else:
        turn_off.append(temp)
```

#### b.循環語句(for)

* 基本概念

```python
name = 'Michael'
for item in name :
    print(item) #字符串
```

```python
movies = ['驚奇隊長','玩具總動員4','蜘蛛人','安納貝爾']
for item in movies :
    print('今日電影:',item) #list
```

```python
movies = ('驚奇隊長','玩具總動員4','蜘蛛人','安納貝爾')
for item in movies :
    print('今日電影:',item) # tuple
```

```python
movies = {1:'驚奇隊長',
          2:'玩具總動員4',
          3:'蜘蛛人',
          4:'安納貝爾'}
for item in movies :
    print('今日電影:',item) # dist
```

```python
movies = {1:'驚奇隊長',
          2:'玩具總動員4',
          3:'蜘蛛人',
          4:'安納貝爾'}
for item in movies.values() :
    print('今日電影:',item) # dist values
```

```python
movies = {1:'驚奇隊長',
          2:'玩具總動員4',
          3:'蜘蛛人',
          4:'安納貝爾'}
for item in movies.keys() :
    print('今日電影:',item) # dist keys
```

```python
movies = {1:'驚奇隊長',
          2:'玩具總動員4',
          3:'蜘蛛人',
          4:'安納貝爾'}
for item in movies.items() :
    print('今日電影:',item) # dist items
```

* for循環語句的進階操作

```python
movies = {1:'驚奇隊長',
          2:'玩具總動員4',
          3:'蜘蛛人',
          4:'安納貝爾'}
for no , name in movies.items() :
    print('今日電影: {}.{}'.format(no,name)) # dist items
```

```python
for number in range(0, 100, 10) :
    print(number)
```

```python
movies = ['驚奇隊長','玩具總動員4','蜘蛛人','安納貝爾']
for index in range(len(movies)) :
    print(index)
```

```python
for temp in range(20, 38 , 1) :
    print(temp , '度')
    break
```

```python
for temp in range(20, 38 , 1) :
    print(temp , '度')
    if temp > 30:
        print('好熱，快開冷氣')
        break
```

```python
for i in range(10) :
    if i % 2 ==0: 
        continue
    print(i)
```

```python
for temp in range(20, 38 , 1) :  
    if temp > 30:
        continue
    print(temp , '度','涼爽~涼爽~') #不執行continue後的語句
```
